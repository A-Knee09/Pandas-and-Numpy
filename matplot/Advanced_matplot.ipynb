# Colored Scatterplots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("csv/iris.csv")
iris.sample(5)
```

```python
iris["Species"] = iris["Species"].replace({"Iris-setosa":0 , "Iris-versicolor":1, "Iris-virginica":2})
iris.sample(5)
```

```python
plt.scatter(iris["SepalLengthCm"], iris["PetalLengthCm"], marker = "x", c=iris["Species"], cmap = "winter")
plt.xlabel("Sepal Lenght in cm")
plt.ylabel("Petal Length in cm")
plt.title("Iris")
plt.grid()
plt.colorbar()
```

# Plot Size

```python
plt.figure(figsize = (15,7))

plt.scatter(iris["SepalLengthCm"], iris["PetalLengthCm"], marker = "x", c=iris["Species"], cmap = "winter")
plt.xlabel("Sepal Lenght in cm")
plt.ylabel("Petal Length in cm")
plt.title("Iris")
plt.grid()
plt.colorbar()
```

# Annotations

```python
batter = pd.read_csv("csv/batter.csv")
sample = batter.head(100).sample(25, random_state = 25)
sample
```

```python
plt.scatter(sample["avg"], sample["strike_rate"])

# Each dot is a batsman but we can't really tell from this plot
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample["avg"], sample["strike_rate"], s = sample["runs"])

for i in range (sample.shape[0]):
    plt.text(sample["avg"].values[i], sample["strike_rate"].values[i], sample["batter"].values[i])

plt.grid()
    
```

# Horizontal and Vertical Lines

```python
# All players with strike rate greater than 130

plt.figure(figsize=(18,10))
plt.scatter(sample["avg"], sample["strike_rate"], s = sample["runs"])

plt.axhline(130, color = 'red')

for i in range (sample.shape[0]):
    plt.text(sample["avg"].values[i], sample["strike_rate"].values[i], sample["batter"].values[i])

plt.grid()
```

```python
# All players with strike rate greater than 130 and average more than 30

plt.figure(figsize=(18,10))
plt.scatter(sample["avg"], sample["strike_rate"], s = sample["runs"])

plt.title("Average and Strike Rate comparision of Batsman")
plt.xlabel("Average")
plt.ylabel("Strike Rate")

plt.axhline(130, color = 'red')
plt.axhline(140, color = "green")
plt.axvline(30, color = 'red')
for i in range (sample.shape[0]):
    plt.text(sample["avg"].values[i], sample["strike_rate"].values[i], sample["batter"].values[i])

plt.grid()
```

# Subplots

```python
fig, ax = plt.subplots(figsize = (15,6))

ax.scatter(batter["avg"] , batter["strike_rate"] , color =  "red", marker = "+")
ax.set_title("Average and Strike Rate comparision of Batsman")
ax.set_xlabel("Average")
ax.set_ylabel("Strike Rate")

fig.show()
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1, sharex = True, figsize = (10,6))
ax[0].scatter(batter["avg"], batter["strike_rate"], color = "red")
ax[1].scatter(batter["avg"], batter["runs"])

ax[0].set_title("Average vs Strike Rate")
ax[0].set_ylabel("Strike Rate")

ax[1].set_title("Average vs Runs")
ax[1].set_xlabel("Average")
ax[1].set_ylabel("Runs")
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,10))

ax[0,0].scatter(batter["avg"], batter["strike_rate"], color = "red")
ax[0,1].scatter(batter["avg"], batter["runs"])
ax[1,0].hist(batter["avg"])
ax[1,1].hist(batter["runs"])

ax[0,0].set_title("Average vs Strike Rate")
ax[0,0].set_ylabel("Strike Rate")

ax[0,1].set_title("Average vs Runs")
ax[0,1].set_xlabel("Average")
ax[0,1].set_ylabel("Runs")
```

```python
# Another way to subplots

fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)  # 2 rows 1 col and 1st plot
ax1.scatter(batter["avg"], batter["strike_rate"], color = "red")

ax2 = fig.add_subplot(2,1,2)
ax2.scatter(batter["avg"], batter["runs"])
```
