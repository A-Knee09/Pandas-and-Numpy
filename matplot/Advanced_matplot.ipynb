# Colored Scatterplots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("csv/iris.csv")
iris.sample(5)
```

```python
iris["Species"] = iris["Species"].replace({"Iris-setosa":0 , "Iris-versicolor":1, "Iris-virginica":2})
iris.sample(5)
```

```python
plt.scatter(iris["SepalLengthCm"], iris["PetalLengthCm"], marker = "x", c=iris["Species"], cmap = "winter")
plt.xlabel("Sepal Lenght in cm")
plt.ylabel("Petal Length in cm")
plt.title("Iris")
plt.grid()
plt.colorbar()
```

# Plot Size

```python
plt.figure(figsize = (15,7))

plt.scatter(iris["SepalLengthCm"], iris["PetalLengthCm"], marker = "x", c=iris["Species"], cmap = "winter")
plt.xlabel("Sepal Lenght in cm")
plt.ylabel("Petal Length in cm")
plt.title("Iris")
plt.grid()
plt.colorbar()
```

# Annotations

```python
batter = pd.read_csv("csv/batter.csv")
sample = batter.head(100).sample(25, random_state = 25)
sample
```

```python
plt.scatter(sample["avg"], sample["strike_rate"])

# Each dot is a batsman but we can't really tell from this plot
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample["avg"], sample["strike_rate"], s = sample["runs"])

for i in range (sample.shape[0]):
    plt.text(sample["avg"].values[i], sample["strike_rate"].values[i], sample["batter"].values[i])

plt.grid()
    
```

# Horizontal and Vertical Lines

```python
# All players with strike rate greater than 130

plt.figure(figsize=(18,10))
plt.scatter(sample["avg"], sample["strike_rate"], s = sample["runs"])

plt.axhline(130, color = 'red')

for i in range (sample.shape[0]):
    plt.text(sample["avg"].values[i], sample["strike_rate"].values[i], sample["batter"].values[i])

plt.grid()
```

```python
# All players with strike rate greater than 130 and average more than 30

plt.figure(figsize=(18,10))
plt.scatter(sample["avg"], sample["strike_rate"], s = sample["runs"])

plt.title("Average and Strike Rate comparision of Batsman")
plt.xlabel("Average")
plt.ylabel("Strike Rate")

plt.axhline(130, color = 'red')
plt.axhline(140, color = "green")
plt.axvline(30, color = 'red')
for i in range (sample.shape[0]):
    plt.text(sample["avg"].values[i], sample["strike_rate"].values[i], sample["batter"].values[i])

plt.grid()
```

# 3D Surface Plots

```python
fig = plt.figure(figsize = (10,7))

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)

z = xx**2 + yy**2



ax = plt.subplot(projection="3d")

p = ax.plot_surface(xx,yy,z,cmap="cividis")
fig.colorbar(p)
```

```python
fig = plt.figure(figsize = (10,7))
z = np.sin(xx) + np.cos(yy)



ax = plt.subplot(projection="3d")

p = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

# Contour Plot

Projecting the 3D plot to a 2D space

```python
z = xx**2 + yy**2
fig = plt.figure(figsize = (10,7))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
z = xx**2 + yy**2
fig = plt.figure(figsize = (10,7))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize = (10,7))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap="cividis")
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize = (10,7))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap="cividis")
fig.colorbar(p)
```

# Heatmap

```python
# On which ball in which over, sixes are performed

delivery = pd.read_csv("csv/IPL_Ball_by_Ball_2008_2022.csv")
delivery.head()
```

```python
temp_df = delivery[(delivery["ballnumber"].isin([1,2,3,4,5,6])) & (delivery["batsman_run"] == 6)]

grid = temp_df.pivot_table(index = "overs", columns = "ballnumber", values = "batsman_run", aggfunc = "count")
```

```python
plt.figure(figsize = (20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()
```
