```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import scipy
```

# 1. Relational Plot
---
- To see the statistical relation between 2 or more variables
- Bivariate Analysis

Plots used : scatterplot and lineplot

## 1.1 Scatterplot

```python
# Scatter plot -> axes level function

tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data = tips, x = "total_bill", y = 'tip', hue="sex", style = 'time', size = 'size')
```

```python
# Scatterplot using figure level function (relplot)
sns.relplot(data = tips , x = "total_bill", y = "tip" , kind = "scatter", hue="sex", style = 'time', size = 'size')
```

## 1.2 Lineplot

```python
# Lineplot
gap = px.data.gapminder()

temp_df = gap[gap["country"] == "India"]
temp_df
```

```python
sns.lineplot(data=temp_df, x = "year", y = "lifeExp")
```

```python
# Using relplot

sns.relplot(data = temp_df, x = "year", y = "lifeExp", kind = "line")
```

```python
# Hue and style

temp_df = gap[gap["country"].isin(["India","Pakistan","China"])]
temp_df
```

```python
sns.relplot(data = temp_df, x = "year", y = "lifeExp", kind = "line", hue = "country")
```

```python
temp_df = gap[gap["country"].isin(["India","Brazil","Germany"])]

sns.relplot(data = temp_df , x = "year", y = "lifeExp", kind = "line", hue = "country", style = "continent")
```

# FacetPlot

Works only with figure level functions

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', kind = 'scatter', col = 'smoker', row = 'sex')
```

```python
# Col wrap

sns.relplot(data = gap, x = "lifeExp", y = "gdpPercap", kind = "scatter", col = "year")
```

```python
sns.relplot(data = gap, x = "lifeExp", y = "gdpPercap", kind = "scatter", col = "year", col_wrap = 3)
```

# 2. Distribution Plots
- Univariate Analysis
- To find out the distribution
- Range of the observation
- Centeral Tendancy
- Is the data bimodal
- Are there outliers

Figure Level : displot

Axes Level : histplot, kdeplot, rugplot

## 2.1 Histplot

```python
# Plotting univariate histogram

sns.histplot(data = tips , x = "tip")
```

```python
sns.displot(data = tips, x = "total_bill", kind = "hist")
```

```python
# bins parameter
sns.displot(data = tips, x = "total_bill", kind = "hist", bins = 20)
```

```python
# Can also be used for categorical data , discrete bins are automatically set for categorical data

# Countplot
sns.displot(data = tips, x = "day", kind = "hist")
```

```python
# Hue parameter
sns.displot(data = tips, x = "tip", kind = "hist", hue = "sex")
```

```python
# element -> step
sns.displot(data = tips, x = "tip", kind = "hist", hue='sex', element = 'step')
```

```python
titanic = sns.load_dataset("titanic")

titanic
```

```python
sns.displot(data = titanic, x = "age", kind = "hist", hue = "sex", element = "step")
```

```python
# Faceting using col and row

sns.displot(data = tips, x = "tip", kind = "hist", col = "sex", element = "step")
```

## 2.2 kdeplot (Kernel Density Estimate)

```python
sns.kdeplot(data = tips, x = "total_bill")
```

```python
# hue -> fill
sns.displot(data=tips, x = "total_bill", hue = "sex", kind = "kde", fill = True)
```

### 2.3 Rugplot
---
A rug plot is a simple, informative graph that shows the distribution of individual data points as small tick marks (or "tassels") along an axis, essentially a 1D scatter plot or a histogram with infinitely narrow bins. It's often used with other plots, like scatter plots or box plots, to reveal underlying data patterns, densities, and clusters, with overlapping marks indicating more data points at a specific value, resembling the fringe on a rug, hence the name

```python
sns.kdeplot(data = tips, x = "total_bill")
sns.rugplot(data = tips, x = "total_bill")
```

```python
# Bivariate Histogram
sns.displot(data = tips, x = "total_bill", y = "tip", kind = "hist")
```

```python
# Bivariate kdeplot

sns.kdeplot(data = tips, x = "total_bill", y = "tip")
```

# 3. Matrix plot
---
- Heatmap
- Clusterplot
  
Data visualization showing relationships between multiple variables in a grid

### 3.1 Heatmap
---
Plots a rectangular data as a color-encoded matrix

```python
gap
```

```python
temp_df = gap.pivot(index = "country", columns="year", values="lifeExp")

# Heatmap can only be done as an axes level
plt.figure(figsize = (15,15))
sns.heatmap(temp_df)
```

```python
# Annot
temp_eu = gap[gap["continent"] == "Europe"].pivot(index = "country", columns = "year", values="lifeExp")

plt.figure(figsize=(15,15))
sns.heatmap(temp_eu, annot=True, linewidth=0.5, cmap="summer")
```

### 3.2 Clustermap
---
plots a hierarchically-clustered heatmap of a given matrix dataset. This visualization reorders the rows and columns of the data matrix to group similar data points near each other, providing a powerful tool for discovering patterns and relationships that might not be obvious in raw data

```python
iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
