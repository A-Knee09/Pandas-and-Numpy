```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Categorical Plots
---

### Categorical Scatter Plots
- Stripplot
- Swarmplot

### Categorical Distribution Plots
- Boxplot
- Violinplot

### Categorical Estimate Plot -> For Central Tendancy
- Barplot
- Pointplot
- Countplot

Figure level function -> `catplot`

```python
# Striplot

tips = sns.load_dataset('tips')
tips
```

```python
#axes level
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False,hue='day')
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',jitter=True,hue='day')
```

```python
# Figure level
sns.catplot(data=tips,x="day",y="total_bill",hue="day",kind="strip")
```

```python
# Jitter
sns.catplot(data=tips,x="day",y="total_bill",hue="sex",kind="strip",jitter=0.2)
```

```python
#swarmplot
sns.catplot(data=tips,x="day",y="total_bill",kind="swarm",hue='sex')
```

# Boxplot
---
A boxplot is a standardized way of displaying the distribution of data based on five number summary ("minimum" first quartile [Q1], median, third quartile[Q3] and "maximum"). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# single boxplot -> numerical col

sns.boxplot(data=tips,y='total_bill')
```

# Violinplot -> Boxplot + KDE

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='day')
```

```python
# barplot

sns.barplot(data=tips, x="sex", y="total_bill", hue='smoker',estimator=np.min)
```

```python
# Pointplot
sns.pointplot(data=tips, x="sex", y="total_bill", hue='smoker',estimator=np.min)
```

```python
# pointplot
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker',errorbar=None)
```

```python
# Countplot
sns.countplot(data=tips,x='sex',hue='day')
```
