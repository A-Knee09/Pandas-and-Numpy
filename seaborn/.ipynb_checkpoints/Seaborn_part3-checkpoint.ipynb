Regression models the relationship between variables so you can explain and predict.

Formally:
You have a dependent variable (Y) and one or more independent variables (X).
Regression learns how Y changes when X changes.

## Why regression exists (the point)

Regression answers questions like:

- If X increases, how much does Y change?

- Is the relationship positive, negative, or zero?

- Can I predict Y for a new X?

- How strong is the relationship?

Plots show patterns.
Regression quantifies them.

````python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
tips.head()
````

````python
# axes level
sns.regplot(data=tips, x = "total_bill", y = "tip")
````

````python
# figure level

sns.lmplot(data=tips,x="total_bill",y="tip",hue='sex')
````

### What is a residual
---
A residual is the error your model makes for a data point.

Formally:

```python
residual = actual value  predicted value
```

If your regression predicts a tip of 3.0 but the actual tip is 3.5:
```python
residual = 3.5  3.0 = +0.5
```

Positive residual  model under-predicted

Negative residual  model over-predicted

Residuals tell you where and how the model is wrong.

### Why residuals matter
---
A regression line always looks nice.
Residuals reveal the truth.

They answer:

- Is a linear model appropriate?

- Is the error random or patterned?

- Does error increase with X?

- Are there outliers?

If residuals show structure, your model is wrong.

### What a residual plot is
---
A residual plot shows:

X-axis: independent variable (or predicted value)

Y-axis: residuals

````python
sns.residplot(data=tips,x="total_bill",y="tip")
````

````python
 # facetplot and facegrid

sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',hue='sex')
````

# Pairplots and Pairgrids
---
A high-level shortcut for pairwise bivariate analysis.

What it does:

Automatically plots every numeric variable vs every other

Diagonal  univariate distributions

Off-diagonal  bivariate relationships

````python
iris = sns.load_dataset("iris")
sns.pairplot(data=iris)

# Histogram -> same variables on both axes
````

````python
g = sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
````

````python
# map_diag -> map_offdiag

g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
````

````python
# map_diag -> map_upper , map_lower

g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
````

````python
# vars

g = sns.PairGrid(data=iris,hue='species',vars=['sepal_length','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
````

# Jointplot
