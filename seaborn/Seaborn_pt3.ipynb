Regression models the relationship between variables so you can explain and predict.

Formally:
You have a dependent variable (Y) and one or more independent variables (X).
Regression learns how Y changes when X changes.

## Why regression exists (the point)

Regression answers questions like:

- If X increases, how much does Y change?

- Is the relationship positive, negative, or zero?

- Can I predict Y for a new X?

- How strong is the relationship?

Plots show patterns.
Regression quantifies them.

```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
tips.head()
```

```python
# axes level
sns.regplot(data=tips, x = "total_bill", y = "tip")
```

```python
# figure level

sns.lmplot(data=tips,x="total_bill",y="tip",hue='sex')
```
